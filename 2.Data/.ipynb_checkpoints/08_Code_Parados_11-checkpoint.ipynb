## Parados 2011
En este notebook procesaremos un .xlsx que contiene el paro registrado por municipios y sexo.  

Obtendremos un DataFrame con la siguiente información:
- Codigo Municipio
- Total Paro Registrado
- Parados Hombres
- Parados Mujeres

```python
import pandas as pd
```

```python
Parados11 = pd.read_excel('08_Input_Parados_11.xlsx',dtype = {'Codigo Municipio' : str})

Parados11.head()
```

Calculamos el total de parados hombres y mujeres

```python
Parados11['Parados Hombres'] = Parados11['Paro hombre edad < 25'] + Parados11['Paro hombre edad 25 -45 '] +\
                              Parados11['Paro hombre edad >=45']

Parados11['Parados Mujeres'] = Parados11['Paro mujer edad < 25'] + Parados11['Paro mujer edad 25 -45 '] +\
                              Parados11['Paro mujer edad >=45']

Parados11.head()
```

```python
Parados11.drop([' Municipio', 
                'Paro hombre edad < 25', 
                'Paro hombre edad 25 -45 ',
                'Paro hombre edad >=45', 
                'Paro mujer edad < 25',
                'Paro mujer edad 25 -45 ', 
                'Paro mujer edad >=45', 
                'Paro Agricultura',
                'Paro Industria', 
                'Paro Construcción', 
                'Paro Servicios',
                'Paro Sin empleo Anterior'], axis = 1, inplace = True)

Parados11.head()
```

```python
Parados11.to_csv('08_Output_Parados_11.csv', header = True, index = False)
```
