## Número de jubilados y estudiantes 2011
En este notebook procesaremos un .xlsx con el número de personas jubiladas y estudiantes desagregado por municipios.  
Obtendremos un DataFrame con la siguiente información:
- Codigo Municipio
- Nº de Jubilados 	
- Nº de Estudiantes

```python
import pandas as pd
```

```python
EstudiantesJubilados11 = pd.read_excel('21_Input_Estudiantes_Jubilados_11.xlsx', usecols = ['Municipio de residencia',
                                                                                            'Jubilado/a, prejubilado/a, pensionista o rentista',
                                                                                            'Estudiantes'], dtype='object')
EstudiantesJubilados11.head()
```

Añadimos el campo Código Municipio al DataFrame asignando el valor de cada fila en función del código que aparece en el campo municipio de residencia.  
Para ello usamos la tabla 01_Output_ProvMun_11.csv que contiene los códigos por municipio:

```python
ProvMun = pd.read_csv('01_Output_ProvMun_11.csv', dtype='object') 
ProvMun.head()
```

```python
MunList = ProvMun['Codigo Municipio'].tolist()
```

```python
def check_list(x):
    for l in MunList:
        if l in x:
            return l
    return ''
```

```python
EstudiantesJubilados11['Codigo Municipio'] = EstudiantesJubilados11['Municipio de residencia'].map(lambda x: check_list(x))
EstudiantesJubilados11.head()
```

```python
EstudiantesJubilados11 = EstudiantesJubilados11.drop(['Municipio de residencia'], axis = 1)
EstudiantesJubilados11 = EstudiantesJubilados11.rename({'Jubilado/a, prejubilado/a, pensionista o rentista': 'Nº de Jubilados',
                                                        'Estudiantes' : 'Nº de Estudiantes'}, axis = 1)
PrimeraColumna = EstudiantesJubilados11.pop('Codigo Municipio')
EstudiantesJubilados11.insert(0,'Codigo Municipio', PrimeraColumna)
EstudiantesJubilados11.head()
```

```python
EstudiantesJubilados11.to_csv('21_Output_Jubliados_Estudiantes_11.csv', header = True, index = False)
```
