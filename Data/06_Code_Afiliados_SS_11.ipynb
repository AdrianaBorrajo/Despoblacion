## Afiliados Seguridad Social 2011
En este notebook procesaremos un .xlsx con el número de afiliados por municipios, dentro del régimen general, agrario, hogar, mar, autónomos, carbón y una última columna con el total.  
Obtendremos un DataFrame con la siguiente información:
- Codigo Municipio
- Regimen General
- Regimen Agrario
- Regimen Hogar
- Regimen Mar
- Regimen Autonomos
- Regimen Carbon
- Total Regimen SS

```python
import pandas as pd
```

```python
AfiliadosSS11 = pd.read_excel('06_Input_Afiliados_SS_11.xlsx',dtype='object')
AfiliadosSS11.head()
```

```python
AfiliadosSS11.drop(['PROVINCIA', 'MUNICIPIO', 'TRAB', 'TOTAL'], axis = 1, inplace = True)
AfiliadosSS11.rename({'CODIGO' : 'Codigo Municipio',
                      'GENERAL': 'Regimen General',
                      'AGRARIO': 'Regimen Agrario', 
                      'HOGAR' : 'Regimen Hogar', 
                      'MAR' : 'Regimen Mar', 
                      'AUTONOMOS' : 'Regimen Autonomos', 
                      'CARBON' : 'Regimen Carbon'}, axis = 1, inplace = True)
AfiliadosSS11.head()
```

Los valores que están entre 0 y 5 se indican en la tabla con <5. Substituimos estos valores por un 2.  
Ademá creamos una nueva columna con la suma de todas las columnas segun régimen.

```python
AfiliadosSS11.dtypes
```

```python
AfiliadosSS11 = AfiliadosSS11.applymap(lambda x: 2 if x == '<5' else x)
```

```python
AfiliadosSS11[['Regimen General', 'Regimen Agrario','Regimen Hogar', 'Regimen Mar','Regimen Autonomos', 'Regimen Carbon']] = AfiliadosSS11[['Regimen General', 
                                                                                                                                            'Regimen Agrario',
                                                                                                                                            'Regimen Hogar', 
                                                                                                                                            'Regimen Mar', 
                                                                                                                                            'Regimen Autonomos', 
                                                                                                                                            'Regimen Carbon']].astype(int)
```

```python
AfiliadosSS11['Total Regimen SS'] = AfiliadosSS11['Regimen General'] + AfiliadosSS11['Regimen Agrario'] + AfiliadosSS11['Regimen Hogar'] + AfiliadosSS11['Regimen Mar'] + AfiliadosSS11['Regimen Autonomos'] + AfiliadosSS11['Regimen Carbon']
```
