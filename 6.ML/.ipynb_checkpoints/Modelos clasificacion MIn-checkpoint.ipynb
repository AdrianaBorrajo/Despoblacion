```python
Despob1acion = pd.read_csv('/home/dsc/Python/Despoblacion/Analisis_Preliminar_Variables/Despoblacion.csv', 
                           dtype = {'Codigo CCAA' : str, 'Codigo Provincia' : str, 'Codigo Municipio' : str})
```

Se dividen los datos, entrenamiento y de test:

```python
X_train, X_test, y_train, y_test = train_test_split(
                                        Despob1acion.drop(['Nombre CCAA', 'Codigo CCAA', 'Nombre Provincia', 'Codigo Provincia', 
                                                   'Nombre Municipio', 'Codigo Municipio', 'Umbral'], 
                                                   axis = 1),
                                        Despob1acion['Umbral'],
                                        train_size   = 0.8,
                                        random_state = 42,
                                        shuffle      = True, 
                                        stratify = Despob1acion['Umbral']
                                    )
```

```python
## Mostramos train y test
```

## Modelado
crearemos un modelo de referencia para cada algoritmo utilizando los parámetros predeterminados establecidos por sklearn y, después de construir los 4 modelos, los compararemos para ver cuál funciona mejor para nuestro caso.

```python
#Definiremos una función de ayuda para evaluar cada modelo entrenado
```

```python
def evaluate_model(model, x_test, y_test):

    # Predict X_test
    y_pred = model.predict(x_test)

    # Calculamos accuracy, precision, recall y f1-score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)

    # Cálculo del area bajo la curva (AUC)
    y_pred_proba = model.predict_proba(x_test)[::,1]
    fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
    auc = metrics.roc_auc_score(y_test, y_pred_proba)

    # Display matriz de confusión
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1,
            'fpr': fpr, 'tpr': tpr, 'auc': auc, 'cm': cm}
```

Entrenamos nuestros 6 modelos:

```python
# Regresión logística 
rl = LogisticRegression(random_state=0)
rl.fit(X_train, y_train)

# Evaluate Model
rl_eval = evaluate_model(rl, X_test, y_test)

# Print result
print('Accuracy:', rl_eval['acc'])
print('Precision:', rl_eval['prec'])
print('Recall:', rl_eval['rec'])
print('F1 Score:', rl_eval['f1'])
print('Area Under Curve:', rl_eval['auc'])
print('Confusion Matrix:')
```

```python
# Maquina de vectores de soporte
svc = SVC(random_state=0, probability=True)
svc.fit(X_train, y_train)

# Evaluate Model
svc_eval = evaluate_model(svc, X_test, y_test)

# Print result
print('Accuracy:', svc_eval['acc'])
print('Precision:', svc_eval['prec'])
print('Recall:', svc_eval['rec'])
print('F1 Score:', svc_eval['f1'])
print('Area Under Curve:', svc_eval['auc'])
print('Confusion Matrix:')
```

```python
# Arbol de decisión 
ad = tree.DecisionTreeClassifier(random_state=0)
ad.fit(X_train, y_train)

# Evaluate Model
ad_eval = evaluate_model(ad, X_test, y_test)

# Print result
print('Accuracy:', ad_eval['acc'])
print('Precision:', ad_eval['prec'])
print('Recall:', ad_eval['rec'])
print('F1 Score:', ad_eval['f1'])
print('Area Under Curve:', ad_eval['auc'])
print('Confusion Matrix:')
```

Comparamos que tan bien funciona cada modelo.  
Para hacer esto, crearemos dos gráficos, primero es un gráfico de barras agrupadas para mostrar el valor de accuracy, precisión, recall y f1 de nuestro modelo, y segundo, un gráfico de líneas para mostrar el AUC de todos nuestros modelos.

```python
# Intitialize figure with two plots
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Model Comparison', fontsize=16, fontweight='bold')
fig.set_figheight(7)
fig.set_figwidth(14)
fig.set_facecolor('white')

# First plot
## set bar size
barWidth = 0.2
dtc_score = [dtc_eval['acc'], dtc_eval['prec'], dtc_eval['rec'], dtc_eval['f1']]
rf_score = [rf_eval['acc'], rf_eval['prec'], rf_eval['rec'], rf_eval['f1']]
nb_score = [nb_eval['acc'], nb_eval['prec'], nb_eval['rec'], nb_eval['f1']]
knn_score = [knn_eval['acc'], knn_eval['prec'], knn_eval['rec'], knn_eval['f1']]

## Posición de las barras en el eje X
r1 = np.arange(len(dtc_score))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

## Make the plot
ax1.bar(r1, dtc_score, width=barWidth, edgecolor='white', label='Decision Tree')
ax1.bar(r2, rf_score, width=barWidth, edgecolor='white', label='Random Forest')
ax1.bar(r3, nb_score, width=barWidth, edgecolor='white', label='Naive Bayes')
ax1.bar(r4, knn_score, width=barWidth, edgecolor='white', label='K-Nearest Neighbors')

## Configure x and y axis
ax1.set_xlabel('Metrics', fontweight='bold')
labels = ['Accuracy', 'Precision', 'Recall', 'F1']
ax1.set_xticks([r + (barWidth * 1.5) for r in range(len(dtc_score))], )
ax1.set_xticklabels(labels)
ax1.set_ylabel('Score', fontweight='bold')
ax1.set_ylim(0, 1)

## Create legend & title
ax1.set_title('Evaluation Metrics', fontsize=14, fontweight='bold')
ax1.legend()

# Segundo plot
## Comparación ROC Curve
ax2.plot(dtc_eval['fpr'], dtc_eval['tpr'], label='Decision Tree, auc = {:0.5f}'.format(dtc_eval['auc']))
ax2.plot(rf_eval['fpr'], rf_eval['tpr'], label='Random Forest, auc = {:0.5f}'.format(rf_eval['auc']))
ax2.plot(nb_eval['fpr'], nb_eval['tpr'], label='Naive Bayes, auc = {:0.5f}'.format(nb_eval['auc']))
ax2.plot(knn_eval['fpr'], knn_eval['tpr'], label='K-Nearest Nieghbor, auc = {:0.5f}'.format(knn_eval['auc']))

## Configure x and y axis
ax2.set_xlabel('False Positive Rate', fontweight='bold')
ax2.set_ylabel('True Positive Rate', fontweight='bold')

## Create legend & title
ax2.set_title('ROC Curve', fontsize=14, fontweight='bold')
ax2.legend(loc=4)

plt.show()
```

Random forest es la opción correcta ya que supera a los otros modelos en 4 de las 5 métricas

Optimizamos el modelo de RF ajustando los hiperparámetros a través de validación cruzada

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [50, 80, 100],
    'max_features': [2, 3, 4],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 300, 500, 750, 1000]
}

# Create a base model
rf_grids = RandomForestClassifier(random_state=0)

# Initiate the grid search model
grid_search = GridSearchCV(estimator=rf_grids, param_grid=param_grid, scoring='recall',
                           cv=5, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

Evaluación del modelo optimizado

```python
# Select best model with best fit
best_grid = grid_search.best_estimator_

# Evaluate Model
best_grid_eval = evaluate_model(best_grid, X_test, y_test)

# Print result
print('Accuracy:', best_grid_eval['acc'])
print('Precision:', best_grid_eval['prec'])
print('Recall:', best_grid_eval['rec'])
print('F1 Score:', best_grid_eval['f1'])
print('Area Under Curve:', best_grid_eval['auc'])
print('Confusion Matrix:\n', best_grid_eval['cm'])
```
