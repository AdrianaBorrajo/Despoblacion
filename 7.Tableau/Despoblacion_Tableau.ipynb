## Fichero Despoblacion para Tableau

En este notebook vamos a añadir al csv Despoblacion, el campo Natcode presente en los ficheros shp correspondientes a península y Canarias.  
Natcode es el campo que se va a utilizar para unir en Tableu, los dos mapas y el fichero Despoblacion.

```python
import geopandas as gpd
import pandas as pd
```

```python
Path_peninsula = 'Recintos_Municipales/recintos_municipales_inspire_peninbal_etrs89.shp'
Peninsula = gpd.read_file(Path_peninsula)

Peninsula.head()
```

```python
Peninsula.crs
```

```python
Path_canarias = 'Recintos_Municipales/recintos_municipales_inspire_canarias_wgs84.shp'
Canarias = gpd.read_file(Path_canarias)

Canarias.head()
```

```python
Canarias.crs
```

Convertimos los puntos de los mapas a un mismo sistema de referencia:

```python
Canarias = Canarias.to_crs(Peninsula.crs)
```

Unimos los dos ficheros:

```python
Mapa = pd.concat([Peninsula, Canarias])
Mapa['Codigo Municipio'] = Mapa['NATCODE'].str.slice(start=6)
Mapa = Mapa[['NATCODE', 'Codigo Municipio']]

Mapa.head()
```

```python
Despoblacion = pd.read_csv('/home/dsc/Python/Despoblacion/5.Analisis_Preliminar_Variables/Despoblacion.csv', 
                           dtype = {'Codigo CCAA' : str, 'Codigo Provincia' : str, 'Codigo Municipio' : str})

Despoblacion.head()
```

Al fichero Despoblacion le unicmos el campo NATCODE con el que luego uniremos ambos ficheros en Tableau:

```python
Despoblacion = Despoblacion.merge(Mapa, on = 'Codigo Municipio')

Despoblacion.head()
```

```python
Despoblacion.to_csv('Despoblacion.csv', header = True, index = False)
```
