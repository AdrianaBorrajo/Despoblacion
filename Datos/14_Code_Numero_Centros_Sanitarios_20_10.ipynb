## Numero centros sanitarios 2020 y 2010
En este notebook procesaremos dos .xlsx; uno con el listado detallado de centros de salud y otro con el listado de hospialtes. Ambos desagregados por municipios.   
Cada uno de los .xlsx tiene dos hojas, una con los datos de 2020 y otra con los de 2010.

```python
import pandas as pd
```

### Numero de centros sanitarios 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Nº Centros Sanitarios

Procesamos el .xlsx de hospitales y de centros de salud del año 2020 y concatenamos:

```python
NumeroHospitales20 = pd.read_excel('14_Input_Numero_Hospitales_20_10.xlsx', usecols = ['Nombre Centro', 'Cód. Municipio'], sheet_name ='2020', dtype='object')
NumeroHospitales20.head()
```

```python
NumeroCentrosSalud20 = pd.read_excel('14_Input_Numero_Centros_Salud_20_10.xlsx', usecols = ['      NOMBRE       ', '       CODMU       '], sheet_name ='2020', dtype='object')
NumeroCentrosSalud20.head()
```

```python
NumeroCentrosSalud20.rename({'      NOMBRE       ' : 'Nombre Centro', '       CODMU       ' : 'Cód. Municipio'}, axis = 1, inplace = True)
```

```python
NumeroCentrosSanitarios20 = pd.concat([NumeroHospitales20, NumeroCentrosSalud20], ignore_index = True)
NumeroCentrosSanitarios20.head()
```

Contamos cuántos centros sanitarios hay por municipio agrupando por Cód. Municipio:

```python
NumeroCentrosSanitarios20 = NumeroCentrosSanitarios20.groupby('Cód. Municipio')['Nombre Centro'].count()
NumeroCentrosSanitarios20 = NumeroCentrosSanitarios20.reset_index()
NumeroCentrosSanitarios20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con NumeroCentrosSanitarios20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
NumeroCentrosSanitarios20.rename({'Cód. Municipio' : 'Codigo Municipio', 'Nombre Centro' : 'Nº Centros Sanitarios'}, axis = 1, inplace = True )
NumeroCentrosSanitarios20.head()
```

```python
NumeroCentrosSanitarios20 = ProvMun.merge(NumeroCentrosSanitarios20 , on='Codigo Municipio')
NumeroCentrosSanitarios20.head()
```

```python
NumeroCentrosSanitarios20['Y2020'] = 2020
PrimeraColumna = NumeroCentrosSanitarios20.pop('Y2020')
NumeroCentrosSanitarios20.insert(0,'Y2020', PrimeraColumna)
NumeroCentrosSanitarios20.head()
```

```python
NumeroCentrosSanitarios20.to_excel('14_Output_Numero_Centros_Sanitarios_20.xlsx', header = True, index = False)
```
