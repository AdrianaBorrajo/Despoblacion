### Explotaciones ganaderas 2020

En este notebook procesaremos un JSON con el número de explotaciones bovinas, ovinas, porcinas y aves desagregado por municipios.

```python
import pandas as pd
import requests
```

```python
path = 'https://servicios.ine.es/wstempus/js/es/DATOS_TABLA/52076?tip=AM&'
json_request = requests.get(path).json()
```

```python
CodBoniList = list()
NumBobiList = list()
CodOviList = list()
NumOviList = list()
CodPorciList = list()
NumPorciList = list()
CodAveList = list()
NumAveList = list()
```

```python
for dato in json_request: 
    for i in range(0,len(dato['Data'])):
        if 'Bovinos, Nº explotaciones' in dato['Nombre']:
            CodBoniList.append(dato['MetaData'][0]['Codigo'])
            NumBobiList.append(dato['Data'][i]['Valor'])
        elif 'Ovino y caprino, Nº explotaciones' in dato['Nombre']:
            CodOviList.append(dato['MetaData'][0]['Codigo'])
            NumOviList.append(dato['Data'][i]['Valor'])
        elif 'Porcinos, Nº explotaciones' in dato['Nombre']:
            CodPorciList.append(dato['MetaData'][0]['Codigo'])
            NumPorciList.append(dato['Data'][i]['Valor']) 
        elif 'Aves de corral, Nº explotacione' in dato['Nombre']:
            CodAveList.append(dato['MetaData'][0]['Codigo'])
            NumAveList.append(dato['Data'][i]['Valor'])            
```

```python
ExplotacionesBovinas20 = pd.DataFrame({
        'Codigo Municipio' : CodBoniList,
        'Nº Explotaciones Bovinas': NumBobiList
    })
ExplotacionesOvinas20 = pd.DataFrame({
        'Codigo Municipio' : CodOviList,
        'Nº Explotaciones Ovinas': NumOviList
    })
ExplotacionesPorcinas20 = pd.DataFrame({
        'Codigo Municipio' : CodPorciList,
        'Nº Explotaciones Procinas': NumPorciList
    })
ExplotacionesAves20 = pd.DataFrame({
        'Codigo Municipio' : CodAveList,
        'Nº Explotaciones Aves': NumAveList
    })
```
