## Uso del suelo 2020 y 2010

En este notebook procesaremos un .xlsx con el número de hectáreas de uso del suelo; industrial, residencial, terciario y otros, desagregao por municipios.  
El .xlsx tiene dos hojas, una con los datos de 2020 y otra con los de 2010.

```python
import pandas as pd
import numpy as np
```

### Uso del suelo 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Otros Usos
- Uso Industrial
- Uso Residelcial
- Uso Terciario

Procesamos la hoja .xlsx correspondiente al año 2020:

```python
UsoSuelo20 = pd.read_excel('20_Input_Uso_Suelo_20_10.xlsx', sheet_name = '2020', dtype='object')
UsoSuelo20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con UsoSuelo20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
UsoSuelo20.rename({'Código':'Codigo Municipio',
                  'Otros usos 2020' : 'Otros Usos',
                  'Uso industrial, almacenamiento y estacionamiento 2020' : 'Uso Industrial',
                  'Uso residencial 2020' : 'Uso Residelcial',
                  'Uso terciario 2020' : 'Uso Terciario'}, axis = 1, inplace = True)
```

```python
UsoSuelo20 = ProvMun.merge(UsoSuelo20, on='Codigo Municipio')
UsoSuelo20.head()
```

Eliminamos la variable Etiqueta e incluimos una columna con el año 2020:

```python
UsoSuelo20.drop(['Etiqueta'], axis = 1, inplace = True)
UsoSuelo20['Y2020'] = 2020
PrimeraColumna = UsoSuelo20.pop('Y2020')
UsoSuelo20.insert(0,'Y2020', PrimeraColumna)
UsoSuelo20.head()
```

Algunas variables toman el valor: N/A - resultado no disponible.  
Lo substituimos por NaN:

```python
UsoSuelo20.replace('N/A - resultado no disponible', np.nan, inplace = True)
```

```python
UsoSuelo20.to_excel('20_Output_Uso_Suelo_20.xlsx', header = True, index = False)
```
