### Numero Empresas 2020

En este notebook procesaremos un .xlsx con el número de empresas desagregado por municipios y tipo de actividad.  
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Empresas Industria
- Empresas Construccion
- Empresas Comercio, Transporte y Hosteleria
- Empresas Informacion y Comunicacion
- Empresas Financieras y Seguros
- Empresas Inmobiliarias
- Empresas Profesionales y Tecnicas
- Empresas Educacion, Sanidad y Servicios Sociales
- Empresas Otros Servicios Personales
- Resto Empresas Servicios
- Total Empresas

```python
import pandas as pd
```

```python
NumeroEmpresas20 = pd.read_excel('12_Input_Numero_Empresas_20.xlsx', dtype='object')
NumeroEmpresas20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con NumeroEmpresas20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

Para realizar el marge, incluimos en la tabla NumeroEmpresas20 el campo Codigo Municipio:

```python
MunList = ProvMun['Codigo Municipio'].tolist()
```

```python
def check_list(x):
    for l in MunList:
        if l in x:
            return l
    return ''
```

```python
NumeroEmpresas20['Codigo Municipio'] = NumeroEmpresas20['Municipios'].map(lambda x: check_list(x))
```

```python
NumeroEmpresas20 = ProvMun.merge(NumeroEmpresas20, on='Codigo Municipio')
NumeroEmpresas20.head()
```

Creamos una columna con el año y renombramos los nombres de las variables:

```python
NumeroEmpresas20['Y2020'] = 2020
PrimeraColumna = NumeroEmpresas20.pop('Y2020')
NumeroEmpresas20.insert(0,'Y2020', PrimeraColumna)
NumeroEmpresas20.rename({'B_E Industria' : 'Empresas Industria', 
                         'F Construcción':'Empresas Construccion',
                         'Comercio, transporte y hostelería' : 'Empresas Comercio Transporte y Hosteleria',
                         'J Información y comunicaciones' : 'Empresas Informacion y Comunicacion', 
                         'K Actividades financieras y de seguros' : 'Empresas Financieras y Seguros',
                         'L Actividades inmobiliarias' : 'Empresas Inmobiliarias',
                         'Actividades profesionales y técnicas' : 'Empresas Profesionales y Tecnicas',
                         'Educación, sanidad y servicios sociales' : 'Empresas Educacion Sanidad y Servicios Sociales',
                         'Otros servicios personales' : 'Empresas Otros Servicios Personales',
                         'Total servicios' : 'Resto Empresas Servicios'}, axis = 1, inplace = True)
NumeroEmpresas20.head()
```

Añadimos una columna con el Total de Empresas: Sumamos la variable Total(suma de todas las variables menos Total Servicios) más la variable Total Servicios.  
El DataFrame contiene valores en blanco y valores con un punto. Lo substituimos por 0 para poder sumar.  
Cambiamos el formato de str a int.

```python
NumeroEmpresas20 = NumeroEmpresas20.replace([' ', '.'], 0)
NumeroEmpresas20[['Total', 'Resto Empresas Servicios']] = NumeroEmpresas20[['Total', 'Resto Empresas Servicios']].astype(int)
NumeroEmpresas20['Total Empreas'] = NumeroEmpresas20['Total'] + NumeroEmpresas20['Resto Empresas Servicios']
NumeroEmpresas20.drop(['Municipios', 'Total'], axis = 1, inplace = True)
NumeroEmpresas20.head()
```

```python
NumeroEmpresas20.to_excel('12_Output_Numero_Empresas_20.xlsx', header = True, index = False)
```
