## Numero de conductores, vehículos y su antigüedad 2020
En este notebook procesaremos un .xlsx que contiene el número de conductores; hombres y mujeres, número de vehículos y su antigüedad media.  
Los datos están desagregados por municipios.  
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Conductores Hombres
- Conductoras Mujeres
- Censo Conductores
- Parque Total
- Antigüedad Media

```python
import pandas as pd
```

```python
Vehiculos20 = pd.read_excel('21_Input_Vehiculos_20.xlsx', 
                                    usecols = ['Código INE',
                                               'Conductores Hombres', 
                                               'Conductoras Mujeres', 
                                               'Censo Conductores', 
                                              'Parque Total',
                                              'Antigüedad Media'],
                                    dtype='object')
Vehiculos20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con Vehiculos20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
Vehiculos20.rename({'Código INE' : 'Codigo Municipio'}, axis = 1, inplace = True )
Vehiculos20.head()
```

```python
Vehiculos20 = ProvMun.merge(Vehiculos20 , on='Codigo Municipio')
Vehiculos20.head()
```

Añadimos una columna con el año 2020:

```python
Vehiculos20['Y2020'] = 2020
PrimeraColumna = Vehiculos20.pop('Y2020')
Vehiculos20.insert(0,'Y2020', PrimeraColumna)
Vehiculos20.head()
```

```python
Vehiculos20.to_excel('21_Output_Vehiculos_20.xlsx', header = True, index = False)
```
