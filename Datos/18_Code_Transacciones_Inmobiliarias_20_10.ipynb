## Número de transacciones inmobiliarias 2020 y 2010
En este notebook procesaremos un .xlsx con el número de transacciones inmobiliarias de viviendas desagregado por municipios.  
El .xlsx tiene dos hojas, una con los datos de 2020 y otra con los de 2010.

### Numero de transacciones inmobiliarias 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Nº Transacciones Inmobiliarias   

Procesamos la hoja .xlsx correspondiente al año 2020:

```python
import pandas as pd
import numpy as np
```

```python
NumeroTransacciones20 = pd.read_excel('18_Input_Transacciones_Inmobiliarias_20_10.xlsx', sheet_name = '2020', dtype='object')
NumeroTransacciones20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con NumeroTransacciones20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
NumeroTransacciones20.rename({'Código' : 'Codigo Municipio', 
                              'Número de transacciones anuales 2020' : 'Nº de transacciones'}, axis = 1, inplace = True)
```

```python
NumeroTransacciones20 = ProvMun.merge(NumeroTransacciones20, on='Codigo Municipio')
NumeroTransacciones20.head()
```

Eliminamos la variable Etiqueta e incluimos una columna con el año 2020:

```python
NumeroTransacciones20.drop(['Etiqueta'], axis = 1, inplace = True)
NumeroTransacciones20['Y2020'] = 2020
PrimeraColumna = NumeroTransacciones20.pop('Y2020')
NumeroTransacciones20.insert(0,'Y2020', PrimeraColumna)
NumeroTransacciones20.head()
```

Algunas variables toman el valor: N/A - resultado no disponible.  
Lo substituimos por NaN:

```python
NumeroTransacciones20.replace('N/A - resultado no disponible', np.nan, inplace = True)
```

```python
NumeroTransacciones20.to_excel('18_Output_Numero_Transacciones_20.xlsx', header = True, index = False)
```
