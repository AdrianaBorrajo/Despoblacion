## Edad medida de los edificios 2020 y 2010
En este notebook procesaremos un .xlsx con la edad media de los edificios por municipios.  
El .xlsx tiene dos hojas, una con los datos de 2020 y otra con los de 2010.


```python
import pandas as pd
import numpy as np
```

### Edad medida de los edificios 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Edad Media Edificios 

Procesamos la hoja .xlsx correspondiente al año 2020:

```python
EdadEdificios20 = pd.read_excel('19_Input_Edad_Edificios_20_10.xlsx', sheet_name = '2020', dtype='object')
EdadEdificios20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con EdadEdificios20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
EdadEdificios20.rename({'Código':'Codigo Municipio', 'Edad media de los edificios 2020' : 'Edad Media Edificios'}, axis = 1, inplace = True)
```

```python
EdadEdificios20 = ProvMun.merge(EdadEdificios20, on='Codigo Municipio')
EdadEdificios20.head()
```

Eliminamos la variable Etiqueta e incluimos una columna con el año 2020:

```python
EdadEdificios20.drop(['Etiqueta'], axis = 1, inplace = True)
EdadEdificios20['Y2020'] = 2020
PrimeraColumna = EdadEdificios20.pop('Y2020')
EdadEdificios20.insert(0,'Y2020', PrimeraColumna)
EdadEdificios20.head()
```

Algunas variables toman el valor: N/A - resultado no disponible.  
Lo substituimos por NaN:

```python
EdadEdificios20.replace('N/A - resultado no disponible', np.nan, inplace = True)
```

```python
EdadEdificios20.to_excel('19_Output_Edad_Edificios_20.xlsx', header = True, index = False)
```
