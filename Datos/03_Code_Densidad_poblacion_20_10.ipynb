## Densidad población 2020 y 2010

En este notebook procesaremos un .xlsx con la superficie de cada uno de los municipios: 03_Input_Densidad_Poblacion_10_20.xlsx 

```python
import pandas as pd
import numpy as np
```

```python
DensidadPoblacion = pd.read_excel('03_Input_Densidad_Poblacion_10_20.xlsx', dtype='object', usecols=('ID_REL',
                                        'PROVINCIA',
                                        'NOMBRE_ACTUAL', 'SUPERFICIE'))
```

```python
DensidadPoblacion.head()
```

```python
DensidadPoblacion.rename({'ID_REL': 'Codigo Municipio'}, axis=1, inplace=True)
```

### Densidad población 2020

Junto con la tabla 02_Output_Poblacion20.xlsx que contiene la población de 2020, obtendremos un Dataframe 
con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Poblacion Total
- Superficie km2
- Densidad Población

```python
Poblacion20 = pd.read_excel('02_Output_Poblacion20.xlsx',dtype='object', usecols=('Y2020',
                                                                                  'Nombre Provincia',
                                                                                  'Codigo Provincia',
                                                                                  'Nombre Municipio', 
                                                                                  'Codigo Municipio', 
                                                                                  'Poblacion Total'))
Poblacion20.head()
```

```python
DensidadPoblacion20 = DensidadPoblacion.merge(Poblacion20, on='Codigo Municipio')
DensidadPoblacion20.head()
```

```python
DensidadPoblacion20 = DensidadPoblacion20.drop(['PROVINCIA', 'NOMBRE_ACTUAL'],axis=1)
```

La superficie está en hectareas, la pasamos a km2 y calculamos la densidad:

```python
DensidadPoblacion20.dtypes
```

```python
DensidadPoblacion20['SUPERFICIE'] = DensidadPoblacion20['SUPERFICIE'].astype(float).round(2)
DensidadPoblacion20['Poblacion Total'] = DensidadPoblacion20['Poblacion Total'].astype(float).round(2)

DensidadPoblacion20['Superficie Km2'] = DensidadPoblacion20['SUPERFICIE'].map(lambda x: x/100).round(2)
DensidadPoblacion20['Densidad Poblacion'] = (DensidadPoblacion20['Poblacion Total'] / DensidadPoblacion20['Superficie Km2']).round(2)
```

```python
ColNue = ['Y2020','Nombre Provincia', 'Codigo Provincia', 'Nombre Municipio','Codigo Municipio',
          'Poblacion Total', 'Superficie Km2', 'Densidad Poblacion']
DensidadPoblacion20 = DensidadPoblacion20[ColNue]
```

```python
DensidadPoblacion20.head()
```

```python
DensidadPoblacion20.to_excel('03_Output_Densidad_Poblacion_20.xlsx', header = True, index = False)
```
