## Numero de centros educativos 2020
En este notebook procesaremos un .xlsx  que contiene el listado detallado de centros educativos desagregados por municipios.  
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Nº Centros Educativos


```python
import pandas as pd
```

```python
CentrosEducativos20 = pd.read_excel('15_Input_Centros_Educativos_20.xlsx', 
                                    usecols = ['Denominación genérica', 'codigo ine'],
                                    dtype='object')
CentrosEducativos20.head()
```

Contamos cuántos centros educativos hay por municipio agrupando por Codigo ine:

```python
CentrosEducativos20 = CentrosEducativos20.groupby('codigo ine')['Denominación genérica'].count()
CentrosEducativos20 = CentrosEducativos20.reset_index()
CentrosEducativos20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con CentrosEducativos20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
CentrosEducativos20.rename({'codigo ine' : 'Codigo Municipio', 'Denominación genérica' : 'Nº Centros Educativos'}, axis = 1, inplace = True )
CentrosEducativos20.head()
```

```python
CentrosEducativos20 = ProvMun.merge(CentrosEducativos20 , on='Codigo Municipio')
CentrosEducativos20.head()
```

```python
CentrosEducativos20['Y2020'] = 2020
PrimeraColumna = CentrosEducativos20.pop('Y2020')
CentrosEducativos20.insert(0,'Y2020', PrimeraColumna)
CentrosEducativos20.head()
```

```python
CentrosEducativos20.to_excel('15_Output_Centros_Educativos_20.xlsx', header = True, index = False)
```
