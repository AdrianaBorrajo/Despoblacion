### Parados 2020 y 2010
En este notebook procesaremos dos Excel, uno para cada uno de los años.  
Estos archivos contienen el paro registrado, por municipios, sexo, edad; <25, 25-45, >45, y sector de actividad económica; agricultura, industria, construcción, servicios y parados de larga duración.

```python
import pandas as pd
```

### Parados 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Paro hombre edad < 25
- Paro hombre edad 25-45
- Paro hombre edad >=45
- Paro mujer edad < 25
- Paro mujer edad 25-45
- Paro mujer edad >=45
- Paro Agricultura
- Paro Industria
- Paro Construcción
- Paro Servicios
- Paro Sin empleo Anterior

```python
Parados20 = pd.read_excel('09_Input_Parados_20.xls',dtype='object')
Parados20.head()
```

```python
Parados20.drop(['Código de CA', 'Comunidad Autónoma', 'Codigo Provincia', 'Provincia', ' Municipio'], axis=1, inplace=True)
```

Cargamos la tabla 01_Output_ProvMun_20.xlsx para posteriormente realizar el merge con Parados20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
Parados20 = Parados20.merge(ProvMun, on='Codigo Municipio')
Parados20.head()
```

```python
Parados20 = Parados20.rename({'Año': 'Y2020'}, axis = 1)
ColNue = ['Y2020','Nombre Provincia', 'Codigo Provincia', 'Nombre Municipio','total Paro Registrado', 
          'Paro hombre edad < 25', 'Paro hombre edad 25 -45 ','Paro hombre edad >=45', 'Paro mujer edad < 25', 
          'Paro mujer edad 25 -45 ', 'Paro mujer edad >=45', 'Paro Agricultura','Paro Industria', 'Paro Construcción', 'Paro Servicios', 
          'Paro Sin empleo Anterior']
Parados20 = Parados20[ColNue]
Parados20.head()
```

```python
Parados20.to_excel('09_Output_Parados_20.xlsx', header = True, index = False)
```
