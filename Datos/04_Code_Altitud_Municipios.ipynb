## Altitud municipios

En este notebook procesaremos el archivo 03_Input_Densidad_Poblacion_10_20.xlsx para extraer la variable altitud de los municipios.  
Obtendremos un Dataframe con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Altitud

```python
import pandas as pd
import numpy as np
```

```python
DensidadPoblacion = pd.read_excel('03_Input_Densidad_Poblacion_10_20.xlsx', dtype='object', usecols=('ID_REL',
                                        'PROVINCIA',
                                        'NOMBRE_ACTUAL', 'ALTITUD'))
```

```python
DensidadPoblacion.head()
```

```python
DensidadPoblacion.rename({'ID_REL': 'Codigo Municipio'}, axis=1, inplace=True)
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con DensidadPoblacion:

```python
ProvMun = pd.read_excel('01_Output_ProvMun20.xlsx', dtype='object') 
ProvMun.head()
```

```python
AltitudMunicipios20 = DensidadPoblacion.merge(ProvMun, on='Codigo Municipio')
AltitudMunicipios20.head()
```

```python
AltitudMunicipios20.rename({'ALTITUD':'Altitud'}, axis=1, inplace=True)
AltitudMunicipios20['Y2020'] = 2020
ColNue = ['Y2020', 'Nombre Provincia', 'Codigo Provincia', 'Nombre Municipio', 'Codigo Municipio', 'Altitud']
AltitudMunicipios20 = AltitudMunicipios20[ColNue]
AltitudMunicipios20.head()
```

```python
AltitudMunicipios20.to_excel('04_Output_Altitud_Municipios_20.xlsx', header = True, index = False)
```
