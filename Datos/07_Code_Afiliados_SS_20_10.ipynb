## Afiliados Seguridad Social 2020 y 2010
En este notebook procesaremos dos Excel, uno para cada uno de los años.  
Estos archivos contienen el número de afiliados dentro del régimen general, agrario, hogar, mar, autónomos, carbón y una última columna con el total.


```python
import pandas as pd
import numpy as np
```

### Afiliados Seguridad Social 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Regimen General
- Regimen Agrario
- Regimen Hogar
- Regimen MaR
- Regimen Autonomos
- Regimen Carbon
- Total Regimen SS

```python
AfiliadosSS20 = pd.read_excel('07_Input_Afiliados_SS_20.xlsx',dtype='object')
AfiliadosSS20.head()
```

Cargamos la tabla 01_Output_ProvMun_20.xlsx para posteriormente realizar el merge con AfiliadosSS20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
AfiliadosSS20.rename({'CODIGO':'Codigo Municipio'}, axis=1, inplace=True)
```

```python
AfiliadosSS20 = AfiliadosSS20.merge(ProvMun, on='Codigo Municipio')
AfiliadosSS20.head()
```

```python
AfiliadosSS20['Y2020'] = 2020
AfiliadosSS20 = AfiliadosSS20.rename({'Reg. General': 'Regimen General',
                                     'R. G.- S.E.Agrario': 'Regimen Agrario', 
                                     'R. G.- S.E.Hogar' : 'Regimen Hogar', 
                                     'R. E. MAR ' : 'Regimen Mar', 
                                     'R. E. T. Autónomos' : 'Regimen Autonomos', 
                                     'R. E. M. Carbón' : 'Regimen Carbon'}, axis = 1)
ColNue = ['Y2020','Nombre Provincia', 'Codigo Provincia', 'Nombre Municipio', 'Codigo Municipio', 'Regimen General', 
          'Regimen Agrario', 'Regimen Hogar','Regimen Mar', 'Regimen Autonomos', 'Regimen Carbon']
AfiliadosSS20 = AfiliadosSS20[ColNue]
AfiliadosSS20.head()
```

Los valores que están entre 0 y 5 se indican en la tabla con <5. Substituimos estos valores por un 2.  
Ademá creamos una nueva columna con la suma de todas las columnas segun régimen.

```python
AfiliadosSS20.dtypes
```

```python
AfiliadosSS20 = AfiliadosSS20.applymap(lambda x: 2 if x == '<5' else x)
```

```python
AfiliadosSS20[['Regimen General', 'Regimen Agrario','Regimen Hogar', 'Regimen Mar','Regimen Autonomos', 'Regimen Carbon']] = AfiliadosSS20[['Regimen General', 'Regimen Agrario','Regimen Hogar', 'Regimen Mar','Regimen Autonomos', 'Regimen Carbon']].astype(int)
```

```python
AfiliadosSS20['Total Regimen SS'] = AfiliadosSS20['Regimen General'] + AfiliadosSS20['Regimen Agrario'] + AfiliadosSS20['Regimen Hogar'] + AfiliadosSS20['Regimen Mar'] + AfiliadosSS20['Regimen Autonomos'] + AfiliadosSS20['Regimen Carbon']
```

```python
AfiliadosSS20.head()
```

```python
AfiliadosSS20.to_excel('07_Output_Afiliados_SS_20.xlsx', header = True, index = False)
```
