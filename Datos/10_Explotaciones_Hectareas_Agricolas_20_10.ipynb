## Explotaciones y hectáreas agrícolas 2020 y 2010
En este notebook procesaremos dos JSON, uno por cada uno de los años, con el número de explotaciones y hectáreas agrícolas desagregado por municipios.

```python
import pandas as pd
import requests
```

### Explotaciones y hectáreas agrícolas 2020
Obtendremos un Dataframe con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Nº Explotaciones Agricolas
- Nº Hectareas Agricolas

```python
path = 'https://servicios.ine.es/wstempus/js/es/DATOS_TABLA/52071?tip=AM&'
json_request = requests.get(path).json()
```

```python
CodExpList = list()
NumExpList = list()
CodSuperList = list()
NumSuperList = list()
```

```python
for dato in json_request: 
    for i in range(0,len(dato['Data'])):
        if 'SAU, Nº explotaciones' in dato['Nombre']:
            CodExpList.append(dato['MetaData'][0]['Codigo'])
            NumExpList.append(dato['Data'][i]['Valor'])
        elif 'SAU, Superficie (ha.)' in dato['Nombre']:
            CodSuperList.append(dato['MetaData'][0]['Codigo'])
            NumSuperList.append(dato['Data'][i]['Valor'])
```

```python
ExplotacionesGanaderas20 = pd.DataFrame({
        'Codigo Municipio' : CodExpList,
        'Nº Explotaciones Ganaderas': NumExpList
    })
HectareasGanaderas20 = pd.DataFrame({
        'Codigo Municipio' : CodSuperList,
        'Nº Hectareas Ganaderas': NumSuperList
    })
```

```python
ExplotacionesGanaderas20.head()
```

```python
HectareasGanaderas20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con ambos DataFrame:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
ExplotacionesGanaderas20 = ProvMun.merge(ExplotacionesGanaderas20, on='Codigo Municipio')
HectareasGanaderas20 = ProvMun.merge(HectareasGanaderas20, on='Codigo Municipio')
```

```python
ExplotacionesGanaderas20['Y2020'] = 2020
PrimeraColumna = ExplotacionesGanaderas20.pop('Y2020')
ExplotacionesGanaderas20.insert(0,'Y2020', PrimeraColumna)
```

```python
ExplotacionesGanaderas20.head()
```

```python
HectareasGanaderas20['Y2020'] = 2020
PrimeraColumna = HectareasGanaderas20.pop('Y2020')
HectareasGanaderas20.insert(0,'Y2020', PrimeraColumna)
```

```python
HectareasGanaderas20.head()
```

```python
ExplotacionesGanaderas20.to_excel('10_Output_Explotaciones_Ganaderas_20.xlsx', header = True, index = False)
HectareasGanaderas20.to_excel('10_Output_Hectareas_Ganaderas_20.xlsx', header = True, index = False)
```
