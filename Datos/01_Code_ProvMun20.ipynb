## Códigos provincias y municipios 2020
En  este notebook procesaremos un .xlsx con la relación de provincias y municipios con sus códigos del INE.\
Obtendremos un Dataframe con la siguiente información:
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio: se compone de cinco dígitos, los dos primeros corresponden al código de la provincia y los tres restantes al del municipio dentro de ésta

```python
import pandas as pd
```

```python
ProvMun = pd.read_excel('01_Input_ProvMun20.xlsx', dtype='object') 

ProvMun['Codigo Municipio'] = ProvMun['CPRO'].map(str)+ProvMun['CMUN'].map(str)

ProvMun = ProvMun.drop(['CODAUTO','CMUN','DC'], axis=1)

ProvMun.rename({'CPRO': 'Codigo Provincia', 
                'NOMBREPRO':'Nombre Provincia',
                'NOMBREMUN': 'Nombre Municipio' }, axis=1, inplace=True)

ProvMun = ProvMun.reindex(columns=['Nombre Provincia','Codigo Provincia','Nombre Municipio','Codigo Municipio'])
```

```python
ProvMun.to_excel('01_Output_ProvMun20.xlsx', header = True, index = False)
```
