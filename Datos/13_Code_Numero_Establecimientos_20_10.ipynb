## Numero establecimientos 2020 y 2010
En este notebook procesaremos un .xlsx con el número de establecimientos desagregado por municipios y por tipos de negocio: agricultura, industria, construcción y servicios.  
El .xlsx tiene dos hojas, una con los datos de 2020 y otra con los de 2010.

```python
import pandas as pd
import numpy as np
```

### Numero establecimientos 2020
Obtendremos un DataFrame con la siguiente información:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio
- Nº establecimientos Agricultura
- Nº establecimientos Industria
- Nº establecimientos Construccion
- Nº establecimientos Servicios

```python
NumeroEstablecimientos20 = pd.read_excel('13_Input_Numero_Establecimientos_20_10.xlsx', '2020', dtype='object')
NumeroEstablecimientos20.head()
```

Cargamos la tabla 01_Output_ProvMun20.xlsx para posteriormente realizar el merge con NumeroEstablecimientos20:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
NumeroEstablecimientos20.rename({'Código':'Codigo Municipio'}, axis = 1, inplace = True)
NumeroEstablecimientos20=ProvMun.merge(NumeroEstablecimientos20, on='Codigo Municipio')
NumeroEstablecimientos20.head()
```

Añadimos la columna con el año 2020 y renombramos las variables:

```python
NumeroEstablecimientos20['Y2020'] = 2020
PrimeraColumna = NumeroEstablecimientos20.pop('Y2020')
NumeroEstablecimientos20.insert(0,'Y2020', PrimeraColumna)
NumeroEstablecimientos20.rename({'Agricultura' : 'Nº Establecimientos Agricultura',
                                'Industria' : 'Nº Establecimientos Industria',
                                'Construcción ' : 'Nº Establecimientos Construccion', 
                                'Servicios' : 'Nº Establecimientos Servicios'}, axis = 1, inplace = True)
NumeroEstablecimientos20.drop(['Etiqueta'], axis = 1, inplace = True)
NumeroEstablecimientos20.head()
```

Algunas variables toman el valor: N/A - resultado no disponible  
Lo substituimos por NaN:

```python
NumeroEstablecimientos20.replace('N/A - resultado no disponible', np.nan, inplace = True)
```

```python
NumeroEstablecimientos20.to_excel('13_Output_Numero_Establecimientos_20.xlsx', header = True, index = False)
```
