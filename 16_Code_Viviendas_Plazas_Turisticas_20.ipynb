## Numero de viviendas y de plazas turísticas 2020
En este notebook procesaremos un JSON que contiene, para cada municipio, el número de viviendas y plazas turísticas en el año 2020.  
Obtendremos dos DataFrame con la siguiente información común:
- Y2020
- Nombre Provincia
- Codigo Provincia
- Nombre Municipio
- Codigo Municipio

Y en cada uno de de los DataFrame, uno de los siguientes campos:
- Nº Viviendas Turisticas
- Nº Plazas Turísiticas

```python
import pandas as pd
import requests
```

```python
path = 'https://servicios.ine.es/wstempus/js/es/DATOS_TABLA/39363?tip=AM&'
json_request = requests.get(path).json()
```

```python
CodVivList = list()
NumVivList = list()
CodPlaList = list()
NumPlaList = list()
```

```python
for dato in json_request: 
    for i in range(0,len(dato['Data'])):
        if 'Viviendas turísticas' in dato['Nombre'] and dato['Data'][i]['Anyo'] == 2020:
            CodVivList.append(dato['MetaData'][0]['Codigo'])
            NumVivList.append(dato['Data'][i]['Valor'])
        elif 'Plazas. Dato base' in dato['Nombre'] and dato['Data'][i]['Anyo'] == 2020:
            CodPlaList.append(dato['MetaData'][0]['Codigo'])
            NumPlaList.append(dato['Data'][i]['Valor'])
```

```python
ViviendasTuristicas20 = pd.DataFrame({
        'Codigo Municipio' : CodVivList,
        'Nº Viviendas Turisticas': NumVivList
    })
PlazasTuristicas20 = pd.DataFrame({
        'Codigo Municipio' : CodPlaList,
        'Nº Plazas Turisticas': NumPlaList
    })
```

```python
ViviendasTuristicas20.head()
```

```python
PlazasTuristicas20.head()
```

Cargamos la tabla 01_Output_ProvMun_20.xlsx para posteriormente realizar el merge con las tablas anteriores:

```python
ProvMun = pd.read_excel('01_Output_ProvMun_20.xlsx', dtype='object') 
ProvMun.head()
```

```python
ViviendasTuristicas20 = ProvMun.merge(ViviendasTuristicas20, on='Codigo Municipio')
PlazasTuristicas20 = ProvMun.merge(PlazasTuristicas20 , on='Codigo Municipio')
```

Añadimos una columna con el año 2020:

```python
ViviendasTuristicas20['Y2020'] = 2020
PrimeraColumna = ViviendasTuristicas20.pop('Y2020')
ViviendasTuristicas20.insert(0,'Y2020', PrimeraColumna)
ViviendasTuristicas20.head()
```

```python
PlazasTuristicas20['Y2020'] = 2020
PrimeraColumna = PlazasTuristicas20.pop('Y2020')
PlazasTuristicas20.insert(0,'Y2020', PrimeraColumna)
PlazasTuristicas20.head()
```

```python
ViviendasTuristicas20.to_excel('16_Output_Viviendas_Turisticas_20.xlsx', header = True, index = False)
PlazasTuristicas20.to_excel('16_Output_Plazas_Turisticas_20.xlsx', header = True, index = False)
```
