Añadimos dos columnas:
- Umbral Ordinal: según el municipio es 0(<12,5) o 1(>=12,5)
- Umbral Nominal: según el municipio es Despoblado(<12,5) o Poblado(>=12,5)

```python
Densidad21['Umbral Ordinal'] = Densidad21['Densidad Poblacion'].map(lambda x: 0 if x < 12.5 else 1)
Densidad21['Umbral Nominal'] = Densidad21['Densidad Poblacion'].map(lambda x: 'Despoblado' if x < 12.5 else 'Poblado')
```

```python
Densidad21.head()
```

```python
Densidad21['Umbral Ordinal'].value_counts()
```

Hacemos un merge con la tabla Despoblación para incluir la columna Umbral de 2021

```python
Despoblacion = Despob1acion.merge(Densidad21, how = 'left', on = 'Codigo Municipio')
```

```python
Despoblacion.head()
```

```python
Despoblacion.drop(['Superficie Km2_y', 'Densidad Poblacion_y'], axis = 1, inplace = True)
```

```python
Despoblacion.rename({'Superficie Km2_x' : 'Superficie Km2', 
                     'Densidad Poblacion_x' : 'Densidad Poblacion'}, axis = 1, inplace = True)
```

Comparativa població hombre y mujer segmentada por municipios despoblados y municipios poblados

```python
Poblado['porcenhom'] = (Poblado['Poblacion Hombre'] / Poblado['Poblacion Total'])*100
Despoblado['porcenhom'] = (Despoblado['Poblacion Hombre'] / Despoblado['Poblacion Total'])*100
Poblado['porcenmuj'] = (Poblado['Poblacion Mujer'] / Poblado['Poblacion Total'])*100
Despoblado['porcenmuj'] = (Despoblado['Poblacion Mujer'] / Despoblado['Poblacion Total'])*100
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,8))

ax1.hist(Despoblado['porcenhom'], bins = 30, color='green', alpha = .5, label = 'Pobl.Hombre')
ax1.hist(Despoblado['porcenmuj'], bins = 30, alpha = .5, label = 'Pobl.Mujer')

ax2.hist(Poblado['porcenhom'], bins = 30, color='green', alpha = .5, label = 'Pobl.Hombre')
ax2.hist(Poblado['porcenmuj'], bins = 30, alpha = .5, label = 'Pobl.Mujer')


ax1.set_xlabel('Municipios despoblados')
ax2.set_xlabel('Municipios poblados')
ax1.legend(frameon=False)
ax2.legend(frameon=False)

plt.show()
```

Grupos de edades segmentadas por municipios despoblados y municipios poblados

```python
Poblado['porcen16'] = (Poblado['Edad < 16'] / Poblado['Poblacion Total'])*100
Despoblado['porcen16'] = (Despoblado['Edad < 16'] / Despoblado['Poblacion Total'])*100
Poblado['porcen45'] = (Poblado['16 =< Edad =<64'] / Poblado['Poblacion Total'])*100
Despoblado['porcen45'] = (Despoblado['16 =< Edad =<64'] / Despoblado['Poblacion Total'])*100
Poblado['porcen65'] = (Poblado['Edad >= 65'] / Poblado['Poblacion Total'])*100
Despoblado['porcen65'] = (Despoblado['Edad >= 65'] / Despoblado['Poblacion Total'])*100

EdadDespoblado = Despoblado[['porcen16', 'porcen45', 'porcen65']]
EdadPoblado = Poblado[['porcen16', 'porcen45', 'porcen65']]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,8))

box1 = ax1.boxplot(EdadDespoblado, patch_artist=True, widths = 0.5, labels = ['Edad < 16', '16 =< Edad =<64', 'Edad >= 65'])
box2 = ax2.boxplot(EdadPoblado, patch_artist=True, widths = 0.5, labels = ['Edad < 16', '16 =< Edad =<64', 'Edad >= 65'])

colors = ['darkseagreen', 'lightblue', 'tan']
for patch, color in zip(box1['boxes'], colors):
    patch.set_facecolor(color)
for patch, color in zip(box2['boxes'], colors):
    patch.set_facecolor(color)
    
#ax1.set_yscale('log')
#ax2.set_yscale('log')
ax1.set_xlabel('Municipios despoblados')
ax2.set_xlabel('Municipios poblados')


plt.show()
```
